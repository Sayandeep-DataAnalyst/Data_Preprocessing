'Weight gained (grams)' variable transformation :

Probability plot to show data distribution before transformation:

```python
stats.probplot(df['Weight gained (grams)'], dist="norm", plot=pylab );
```

Transforming the data using the log technique :

```python
    
df['Weight gained (grams)'] = np.log(df['Weight gained (grams)'])
```

Probability plot to show data distribution after transformation:

```python
stats.probplot(df['Weight gained (grams)'], dist="norm", plot=pylab );
```
