Importing the necessary packages/dependencies  :

Normalization of the variable through the Sklearn technique :

```python
df['Area'] = preprocessing.scale(df['Area'])
df['Perimeter '] = preprocessing.scale(df['Perimeter '])
df['Compactness'] = preprocessing.scale(df['Compactness'])
df['length'] = preprocessing.scale(df['length'])
df['Assymetry_coeff '] = preprocessing.scale(df['Assymetry_coeff'])
df['len_ker_grove'] = preprocessing.scale(df['len_ker_grove'])
```

Plotting the curve so that we can chck the plotting :

```python
df_plot = df[['Area','Perimeter ','Compactness','length','Width','Assymetry_coeff', 'len_ker_grove', 'Type']].copy()
```
