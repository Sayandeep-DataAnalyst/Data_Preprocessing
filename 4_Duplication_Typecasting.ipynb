Q2. Check for the duplicate values, and handle the duplicate values (ex. drop)

```python
df.InvoiceNo.duplicated().count()
```

```python
df_inv_dup = df[~df.duplicated('StockCode')]
```

```python
df_inv_dup['StockCode'].sort_values()
```

```python
df.sort_values("StockCode", inplace = True)
 
# making a bool series
bool_series = df["StockCode"].duplicated(keep = False)
 
# bool series
bool_series
 
# passing NOT of bool series to see unique values only
df = df[~bool_series]
 
# displaying data
df.info()
df
```
